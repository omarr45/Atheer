```python
from ATHEER_Arabic_Stemmer import AtheerStemmer

string = 'وَقَفَ الطُّلابُ في صُفوفٍ مُرَتَّبَةٍ، ثُمَّ خَرَجوا إلى حَقْلٍ قَرِيبٍ. أَخَذوا مَعَهُمْ غِراسَ التّينِ والخَوْخِ، حَفَرَ الطُّلابُ الأَرْضَ وَزَرَعوا الغِرَاسَ، ثمَّ سَقَوْها المَاءَ. عادَ الطُّلابُ إلى المَدْرَسَةِ مسرورين.'

stems_list = AtheerStemmer.text2stem(string, full=False)
stemP_list = AtheerStemmer.text2stem(string, full=True)
roots_list = AtheerStemmer.text2root(string)
```

```python
import pandas as pd

df = pd.DataFrame(columns=['original', 'stem+', 'stem','root'])
df['original'] = string.split()
df['stem+'] = stemP_list
df['stem'] = stems_list
df['root'] = roots_list

# [::-1] to print from right to left
df[::-1].transpose()
```

```python
df.to_csv('Notebook/output.csv')
df[::-1].transpose().to_csv('Notebook/mirrored-output.csv')
```
